# Parse

```python
import copy
import json
import os

from loguru import logger

from magic_pdf.data.data_reader_writer import FileBasedDataWriter
from magic_pdf.libs.draw_bbox import draw_layout_bbox, draw_span_bbox
from magic_pdf.pipe.OCRPipe import OCRPipe
from magic_pdf.pipe.TXTPipe import TXTPipe
from magic_pdf.pipe.UNIPipe import UNIPipe
```

## File path

```python
temp_pdf_path = 'manual.pdf'
```

```python
!wget https://www.cryptopro.ru/sites/default/files/products/pdf/files/CryptoProPDF_UserGuide.pdf -O {temp_pdf_path}
```

```python
import os

pdf_name = os.path.basename(temp_pdf_path).split('.')[0]
pdf_bytes = open(temp_pdf_path, 'rb').read()
parse_method = 'txt'

# results
output_dir = 'results'
output_path = os.path.join(output_dir, pdf_name)
# assets
output_image_path = os.path.join(output_path, 'images')
image_path_parent = os.path.basename(output_image_path)

pdf_name, output_path, output_image_path, image_path_parent
```

## Choose parse mode (OCR or TXT)

```python
model_json = []
orig_model_list = copy.deepcopy(model_json)
```

```python
image_writer, md_writer = FileBasedDataWriter(output_image_path), FileBasedDataWriter(output_path)
```

```python
if parse_method == 'auto':
    jso_useful_key = {'_pdf_type': '', 'model_list': model_json}
    pipe = UNIPipe(pdf_bytes, jso_useful_key, image_writer, lang='ru')
elif parse_method == 'txt':
    pipe = TXTPipe(pdf_bytes, model_json, image_writer, lang='ru')
elif parse_method == 'ocr':
    pipe = OCRPipe(pdf_bytes, model_json, image_writer, lang='ru')
else:
    logger.error('unknown parse method, only auto, ocr, txt allowed')

pipe
```

```python
pipe = TXTPipe(pdf_bytes, model_json, image_writer, lang='ru')
```

```python
# Classify if parse 'auto'
pipe.pipe_classify()
```

## Run pipeline for each page

```python
pipe.pipe_analyze()
```

```python
pipe.model_list[:2]
```

## Merge predictions

```python
%%time
pipe.pipe_parse()
```

## Serialize result

```python
content_list = pipe.pipe_mk_uni_format(image_path_parent, drop_mode='none')
md_content = pipe.pipe_mk_markdown(image_path_parent, drop_mode='none')
```

```python
def json_md_dump(
        pipe,
        md_writer,
        pdf_name,
        content_list,
        md_content,
        orig_model_list,
):
    # 写入模型结果到 model.json

    md_writer.write_string(
        f'{pdf_name}_model.json',
        json.dumps(orig_model_list, ensure_ascii=False, indent=4)
    )

    # 写入中间结果到 middle.json
    md_writer.write_string(
        f'{pdf_name}_middle.json',
        json.dumps(pipe.pdf_mid_data, ensure_ascii=False, indent=4)
    )

    # text文本结果写入到 conent_list.json
    md_writer.write_string(
        f'{pdf_name}_content_list.json',
        json.dumps(content_list, ensure_ascii=False, indent=4)
    )

    # 写入结果到 .md 文件中
    md_writer.write_string(
        f'{pdf_name}.md',
        md_content,
    )


# 可视化
def draw_visualization_bbox(pdf_info, pdf_bytes, local_md_dir, pdf_file_name):
    # 画布局框，附带排序结果
    draw_layout_bbox(pdf_info, pdf_bytes, local_md_dir, pdf_file_name)
    # 画 span 框
    draw_span_bbox(pdf_info, pdf_bytes, local_md_dir, pdf_file_name)
```

## Store processing mid results vis

```python
json_md_dump(pipe, md_writer, pdf_name, content_list, md_content, orig_model_list)
draw_visualization_bbox(pipe.pdf_mid_data['pdf_info'], pdf_bytes, output_path, pdf_name)
```
